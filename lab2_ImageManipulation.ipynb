```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
```

```python
img = io.imread('/content/drive/MyDrive/ArtificialIntelligence/lab2.jpg')
io.imshow(img)
```

```python
img.shape
```

```python
img.dtype
```

```python
from skimage.transform import resize
resized_img = resize(img, (img.shape[0]//5, img.shape[1]//5), anti_aliasing=True)
plt.imshow(resized_img)
```

```python
from skimage.transform import rotate
rot90_img = rotate(img, 90, resize=True)
rot180_img = rotate(img, 180, resize=True)
fig, axes = plt.subplots(nrows=1, ncols=2) # 12   subplot 
ax = axes.ravel()
ax[0].imshow(rot90_img)
ax[0].set_title('90 degree rotated image')
ax[1].imshow(rot180_img)
ax[1].set_title('180 degree rotated image')
```

```python
crop_img = img[1500:2000, 100:600]
io.imshow(crop_img)
```

```python
from skimage import draw
RI = resize(img, (200,200))
# colors : Red, Green, Blue
colorR = np.array([255,0,0], dtype=np.uint8)
colorG = np.array([0,255,0], dtype=np.uint8)
colorB = np.array([0,0,255], dtype=np.uint8)
# bounding box format (x1, y1, w, h)
BB = (10,50,20,10)
RI[BB[1], BB[0]:BB[0]+BB[2]] = colorR
RI[BB[1]:BB[1]+BB[3], BB[0]] = colorG
RI[BB[1]:BB[1]+BB[3], BB[0]+BB[2]] = colorG
RI[BB[1]+BB[3], BB[0]:BB[0]+BB[2]] = colorB
io.imshow(RI)
```
